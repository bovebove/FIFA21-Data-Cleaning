**Remark 3:**
- We have no information on the start and end date of the contract of players on loan. Indeed, when a player is under contract with club A, this club can loan him to another club B. The real starting date of the player's contract is the one with his home club, i.e. club A, and not the starting date of the loan with club B. Similarly, the true end date of the player's contract is the date with club A and not the loan end date.

- As we do not have the 'Start date' and 'End date' information for players on loan, we will assume that these dates are equal to the loan start date and loan end date respectively.

- Especially since we notice that the number of players on loan (1113) + the number of free players (237) = number of players without a contract end date (1250).

- So we don't need the column "Loan Date End" anymore, which we will delete after assigning its values to the column "End Year".
