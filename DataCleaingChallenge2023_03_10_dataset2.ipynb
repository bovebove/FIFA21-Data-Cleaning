```python
# We use regular expression to work on "Start Year" to tranform it to "Start date"
year_exp = re.compile(r"([0-9]{4})")
date_exp1 = re.compile(r"([0-9]{2}/[0-9]{2}/[0-9]{4})")
date_exp2 = re.compile(r"([0-9]{2}/[0-9]{1}/[0-9]{4})")
```

The real starting date of the player's contract is the one with his home club, i.e. club A, and not the starting date of the loan with club B. Similarly, the true end date of the player's contract is the date with club A and not the loan end date.

- As we do not have the 'Start date' and 'End date' information for players on loan, we will assume that these dates are equal to the loan start date and loan end date respectively.

- Especially since we notice that the number of players on loan (1113) + the number of free players (237) = number of players without a contract end date (1250).
